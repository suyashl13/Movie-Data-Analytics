# Data analytics on movie dataset.
This is data analytics performed on movies dataset.


#### Import Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### 1. Read your CSV file (Dataset) into DataFrame as df

```python
movies_dataset = pd.read_csv('./movie.csv')
movies_dataset.head()
```

```python
pd.Series(movies_dataset.keys())
```

#### 2. Get the total count of Color and Black/White movies

```python
movies_dataset['color'].value_counts()
```

#### 3. Plot bar graph and Pie chart for the above movies count

```python
sns.countplot(x='color',data=movies_dataset)
```

#### 4. Display all the unique director names

```python
movies_dataset['director_name'].value_counts()
```

#### 5. Display all the diffent generes of movie

```python
m_genre = set()
for item in movies_dataset['genres']:
    for i in item.split('|'):
        m_genre.add(i)
            
pd.Series(list(m_genre))
```

#### 6. Get the total count of movies released in New Zealand

```python
movies_dataset['country'].value_counts()['New Zealand']
```

#### 7. List all the movies with all details which were released in New Zealand

```python
movies_dataset.copy().loc[movies_dataset['country'] == 'New Zealand']
```

#### 8. Dispaly the bargraph with count for the movies which are released in Austraial, Germany and Japan

```python
bar_data = {
    'Australia':movies_dataset['country'].value_counts()['Australia'],
    'Germany':movies_dataset['country'].value_counts()['Germany'],
    'Japan':movies_dataset['country'].value_counts()['Japan'],
}

plt.bar(
    bar_data.keys(),
    bar_data.values(),
)
```

#### 9. Get the list of 5 movies with highest budget

```python
movies_dataset.copy().sort_values(by=['budget'], ascending=False).head(5)[['movie_title','budget']]
```

#### 10. Dispaly the list of movies which are released in year 2009

```python
movies_dataset.copy().loc[movies_dataset['title_year'] == 2009][['movie_title','title_year']]
```

#### 11. Dispaly the list of movies which are released in year 2009 and does not belgon to USA

```python
movies_dataset.copy().loc[movies_dataset['title_year'] == 2009].loc[movies_dataset['country'] != 'USA'][['movie_title','title_year', 'country']]
```

#### 12. Get the Count for movies which has duration less than 100 and greater than 100 and plot the bar graph for same

```python
mov_count = {
    'movies_less_that_100' : movies_dataset['duration'].copy().loc[movies_dataset['duration'] < 100].count(),
    'movies_more_that_100' : movies_dataset['duration'].copy().loc[movies_dataset['duration'] > 100].count(),
}
```

```python
plt.bar(
    mov_count.keys(),
    mov_count.values(),
)
```

#### 13. List the top 10 director names for highest director-facebook likes

```python
movies_dataset.copy().sort_values(by=['director_facebook_likes'], ascending=False).head(10)[['movie_title','director_facebook_likes']]
```

#### 14. Display the list of movies for top 5 with movie-facebook likes

```python
top_5_fblike_movies = movies_dataset.copy().sort_values(by=['movie_facebook_likes'], ascending=False).head(5)[['movie_title','movie_facebook_likes']]
top_5_fblike_movies
```

#### 15. Plot the graph for top 5 movie facebook likes

```python
plt.bar(
    top_5_fblike_movies['movie_title'].values,
    top_5_fblike_movies['movie_facebook_likes'].values,
    width=0.5
)
```
